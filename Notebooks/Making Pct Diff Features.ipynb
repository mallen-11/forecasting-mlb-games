```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

import datetime as dt
```

```python
mlb_games = pd.read_csv('../data/saved_datasets/my_training_dataset2.csv')
mlb_games = mlb_games.sort_values(by='date')
```

```python
mlb_games.head()
```

```python
mlb_games.columns
```

```python
mlb_games['home_ops'] = mlb_games['home_slg'] + mlb_games['home_obp']
mlb_games['away_ops'] = mlb_games['away_slg'] + mlb_games['away_obp']
```

```python
mlb_games['home_RD'] = mlb_games['home_R_offset1year'] - mlb_games['home_RA_offset1year']
mlb_games['away_RD'] = mlb_games['away_R_offset1year'] - mlb_games['away_RA_offset1year']
```

```python
mlb_games = mlb_games.replace([np.inf, -np.inf], 1000)
```

```python
# > 0 = home is better
mlb_games['pitcher_WHIP_diff'] = (mlb_games['pitcher_home_WHIP_offset1game'] - mlb_games['pitcher_away_WHIP_offset1game'])
mlb_games['pitcher_ERA_diff'] = (mlb_games['pitcher_home_ERA_offset1game'] - mlb_games['pitcher_away_ERA_offset1game'])
mlb_games['pitcher_IP_diff'] = (mlb_games['pitcher_home_IP_offset1game'] - mlb_games['pitcher_away_IP_offset1game'])
mlb_games['team_WHIP_diff'] = (mlb_games['team_home_WHIP_offset1year'] - mlb_games['team_away_WHIP_offset1year'])
mlb_games['W-L_diff'] = (mlb_games['home_W-L-pct_offset1year'] - mlb_games['away_W-L-pct_offset1year'])
mlb_games['team_ERA_diff'] = (mlb_games['team_home_ERA_offset1year'] - mlb_games['team_away_ERA_offset1year'])
mlb_games['ops_diff'] = mlb_games['home_ops'] - mlb_games['away_ops']
mlb_games['RD_diff'] = mlb_games['home_RD'] - mlb_games['away_RD']
mlb_games['FP_diff'] = mlb_games['home_FP_offset1year'] - mlb_games['away_FP_offset1year']
mlb_games['Rank_diff'] = mlb_games['home_Rank_offset1year'] - mlb_games['away_Rank_offset1year']
mlb_games['WPA_diff'] = mlb_games['pitcher_home_WPA_offset1game'] - mlb_games['pitcher_away_WPA_offset1game']


# scaling doesn't matter for many cases
mlb_games['pitcher_WHIP_pct_diff'] = (mlb_games['pitcher_WHIP_diff'])/(mlb_games['pitcher_home_WHIP_offset1game'])*100
mlb_games['pitcher_ERA_pct_diff'] = (mlb_games['pitcher_ERA_diff'])/(mlb_games['pitcher_home_ERA_offset1game'])*100
mlb_games['pitcher_IP_pct_diff'] = (mlb_games['pitcher_IP_diff'])/(mlb_games['pitcher_home_IP_offset1game'])*100
mlb_games['team_WHIP_diff'] = (mlb_games['team_WHIP_diff'])/(mlb_games['team_home_WHIP_offset1year'])*100
mlb_games['team_ERA_pct_diff'] = (mlb_games['team_ERA_diff'])/(mlb_games['team_home_ERA_offset1year'])*100
mlb_games['ops_pct_diff'] = (mlb_games['ops_diff'] / mlb_games['home_ops'])*100
#mlb_games['attendance_pct_diff'] = (mlb_games['attendance_diff'] / mlb_games['home_Avg_Attendance_offset1year'])*100
mlb_games['RD_pct_diff'] = (mlb_games['RD_diff'] / mlb_games['home_RD'])*100
mlb_games['FP_pct_diff'] = (mlb_games['FP_diff'] / mlb_games['home_FP_offset1year'])*100
mlb_games['Rank_pct_diff'] = (mlb_games['Rank_diff'] / mlb_games['home_Rank_offset1year'])*100
mlb_games['WPA_pct_diff'] = (mlb_games['WPA_diff'] / mlb_games['pitcher_home_WPA_offset1game'])*100
```
