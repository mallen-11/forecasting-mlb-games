```python
teams = ['TOR', 'SFN', 'SEA', 'NYA', 'LAN', 'BAL', 'COL', 'CHN', 'MIA',
       'CLE', 'CIN', 'TEX', 'TBA', 'DET', 'ATL', 'HOU', 'MIL', 'WAS',
       'PHI', 'OAK', 'ARI', 'CHA', 'BOS', 'KCA', 'SLN', 'PIT', 'NYN',
       'MIN', 'SDN', 'ANA']

old_teams = ['LAA', 'FLA', 'WSN', 'SFG', 'NYY', 'NYM', 'CHW', 'CHC', 'KCR', 'SDP', 'LAD',
            'STL', 'TBD']
```

```python
def convert_old_team_ids(team):
    if (team == 'LAA'):
        return 'ANA' # matches up with previous table
    if (team == 'FLA'):
        return 'MIA' # matches up with previous table
    if (team == 'WSN'):
        return 'WAS' # matches up with previous table
    if (team == 'SFG'):
        return 'SFN'
    if (team == 'NYY'):
        return 'NYA'
    if (team == 'NYM'):
        return 'NYN'
    if (team == 'CHW'):
        return 'CHA'
    if (team == 'CHC'):
        return 'CHN'
    if (team == 'KCR'):
        return 'KCA'
    if (team == 'SDP'):
        return 'SDN'
    if (team == 'LAD'):
        return 'LAN'
    if (team == 'STL'):
        return 'SLN'
    if (team == 'TBD'):
        return 'TBA'
    return team
```

```python
def group_teams(team):
    if team in ['LAA', 'ANA']:
        return ['LAA', 'ANA']
    if team in ['FLA', 'MIA']:
        return ['FLA', 'MIA']
    if team in ['WSN', 'WAS']:
        return ['WSN', 'WAS']
    if team in ['SFG', 'ANA']:
        return ['SFG', 'SFN']
    if team in ['NYY', 'NYA']:
        return ['NYY', 'NYA']
    if team in ['NYM', 'NYN']:
        return ['NYM', 'NYN']
    if team in ['CHW', 'CHA']:
        return ['CHW', 'CHA']
    if team in ['CHC', 'CHN']:
        return ['CHC', 'CHN']
    if team in ['KCR', 'KCA']:
        return ['KCR', 'KCA']
    if team in ['SDP', 'SDN']:
        return ['SDP', 'SDN']
    if team in ['LAD', 'LAN']:
        return ['LAD', 'LAN']
    if team in ['STL', 'SLN']:
        return ['STL', 'SLN']
    if team in ['TBD', 'TBA']:
        return ['TBD', 'TBA']
    return [team]
```

```python
def aggregate_teams(team):
    names = group_teams(team)
    team = names[-1]
    summary_dict = dict()
    for y in range(2000, 2020):
        for n in names:
            try:
                year_df = pyb.schedule_and_record(y, n)
                wins = int(year_df.iloc[-1]['W-L'].split('-')[0])
                losses = int(year_df.iloc[-1]['W-L'].split('-')[1])
                win_loss_pct = wins / (wins + losses)
                start_rank = int(float(year_df.iloc[0]['Rank']))
                end_rank = int(float(year_df.iloc[-1]['Rank']))
                summary_dict[y] = [team, wins, losses, win_loss_pct, year_df['Attendance'].mean(), start_rank, end_rank]
            except Exception as e:
                continue

    team_df = pd.DataFrame(summary_dict).transpose()
    team_df = team_df.reset_index()
    team_df.columns = ['Year', 'Tm', 'Wins', 'Losses', 'W-L-pct', 'Avg_Attendance', 'Initial_Rank', 'Final_Rank']
    team_df = team_df.sort_values('Year')
    
    team_df.to_csv(f'../data/team_stats/{team}.csv', index=False)
    print(f'Saved {team}.csv')
```

```python
for t in teams:
    aggregate_teams(t)
```
